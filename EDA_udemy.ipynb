## Clasificacion de Variables

Variables cuantitativas:

course_id: Identificador único del curso. Es una variable cuantitativa discreta.

price: Precio del curso. Es una variable cuantitativa continua.

num_subscribers: Número de suscriptores al curso. Es una variable cuantitativa discreta.

num_reviews: Número de revisiones o comentarios del curso. Es una variable cuantitativa discreta.

num_lectures: Número de conferencias o clases en el curso. Es una variable cuantitativa discreta.

content_duration: Duración del contenido del curso en horas. Es una variable cuantitativa continua.

Variables cualitativas:

course_title: Título del curso. Es una variable cualitativa nominal.

url: URL del curso. Es una variable cualitativa nominal.

is_paid: Indica si el curso es de pago o gratuito. Es una variable cualitativa nominal.

level: Representa el nivel de dificultad del curso. Es una variable cualitativa ordinal.

published_timestamp: Fecha y hora de publicación del curso. Es una variable cualitativa nominal.

subject: Representa el tema o categoría del curso. Es una variable cualitativa nominal.

Variable dicotómica:

is_paid: Indica si el curso es de pago o gratuito. Toma los valores True o False.

Variables polinómicas:

level: Representa el nivel de dificultad del curso. Puede tener múltiples valores como "All Levels", "Beginner", "Intermediate", "Advanced", etc.

subject: Representa el tema o categoría del curso. Puede haber varios valores posibles, como "Business Finance", "Web Development", "Data Science", etc.

Variables discretas:

course_id: Identificador único del curso.

num_subscribers: Número de suscriptores al curso.

num_reviews: Número de revisiones o comentarios del curso.

num_lectures: Número de conferencias o clases en el curso.

Variables continuas:

price: Precio del curso.

content_duration: Duración del contenido del curso en horas.

## Análisis Multivariable

Matrix de correlación:

```python
numeric_columns = ['price', 'num_subscribers', 'num_reviews', 'num_lectures', 'content_duration']
correlation_matrix = udemy[numeric_columns].corr()

plt.figure(figsize=(8, 6))
ax = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de Correlación')

# Rotar las etiquetas del eje x
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

plt.show()
```

Ditribucion de tematicas por nivel de dificultad en relacion a la cantidad de cursos publicados

## Creamos la columna venta miltiplicando price por num_suscribers

```python
udemy['venta'] = udemy['price'] * udemy['num_subscribers']
```

## Top 10 de cursos con mayores ventas en dolares

```python
import matplotlib.ticker as ticker


top_10_sales = udemy.nlargest(10, 'venta')[['course_title', 'venta']].sort_values('venta', ascending=False)

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=top_10_sales, x='venta', y='course_title', palette='viridis')

# Configuración de etiquetas en el eje x
ax.xaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.set_xlabel(' Venta en $')
ax.set_ylabel('Nombre del Curso')
ax.set_title('Top 10 de cursos con mayores ventas')

plt.show()
```

## Nivel de ventas según precio y nivel que son las variables con las que contamos en el data set 

Segmentación por precio:

```python
# Segmentación por precio y ventas
price_segmentation = udemy.groupby('price')['venta'].sum().reset_index()
price_segmentation = price_segmentation.nlargest(15, 'venta')  # Filtrar el top 15 en orden descendente

plt.figure(figsize=(8, 6))
ax = sns.barplot(data=price_segmentation, x='price', y='venta', palette='viridis')

# Configuración de etiquetas en el eje y
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.set_xlabel('Precio $')
ax.set_ylabel('Ventas $')
ax.set_title('Segmentación por precio')

plt.show()
```

Segmentación por nivel:

```python
level_segmentation = udemy.groupby('level')['venta'].sum().reset_index()

plt.figure(figsize=(10, 6))
ax = sns.barplot(data=level_segmentation, x='level', y='venta', palette='viridis')

# Configuración de etiquetas en el eje y
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, p: format(int(x), ',')))
ax.set_xlabel('Nivel de Dificultad')
ax.set_ylabel('Ventas en $')
ax.set_title('Segmentación por nivel')
ax.set_xticklabels(ax.get_xticklabels(), rotation=45)

plt.show()
```

### KPI Udemy

```python

# Calcular el número de inscritos gratuitos y pagados
num_inscritos_gratuitos = len(udemy[udemy['is_paid'] == False])
num_inscritos_pagados = len(udemy[udemy['is_paid'] == True])

# Calcular la tasa de conversión
tasa_conversion = (num_inscritos_pagados / num_inscritos_gratuitos) * 100

# Imprimir el resultado
print(f"Tasa de conversión de inscritos gratuitos a inscritos pagados: {tasa_conversion:.2f}%")

# Visualizar la tasa de conversión en un gráfico de barras mejorado
conversion_data = {'Tipo de Inscritos': ['Gratuitos', 'Pagados'],
                   'Cantidad de Inscritos': [num_inscritos_gratuitos, num_inscritos_pagados]}

# Colores personalizados para los valores del eje y
colores = ['#ff7f0e', '#1f77b4']

fig, ax = plt.subplots(figsize=(8, 6))  # Ajustar el tamaño del gráfico

# Gráfico de barras con colores personalizados
ax.bar(conversion_data['Tipo de Inscritos'], conversion_data['Cantidad de Inscritos'], color=colores)

# Etiquetas y título del gráfico
ax.set_xlabel('Tipo de Inscritos')
ax.set_ylabel('Cantidad de Inscritos')
ax.set_title('Tasa de Conversión de Inscritos Gratuitos a Inscritos Pagados')

# Ajustar la calidad del diseño del gráfico
plt.tight_layout()

plt.show()
```

KPI: Crecimiento de suscriptores en Udemy

Este KPI nos ayudará a medir el crecimiento de suscriptores en Udemy a lo largo del tiempo. Podemos calcularlo de la siguiente manera:

Explicación y funcionalidad del KPI:

Este KPI nos permitirá visualizar y analizar la tendencia del crecimiento de suscriptores en Udemy a lo largo del tiempo. Al observar el gráfico de líneas, podremos identificar los períodos de mayor crecimiento y evaluar el impacto de ciertos eventos o estrategias en el número de suscriptores. También podremos comparar el crecimiento en diferentes años para identificar posibles patrones estacionales o cambios en las estrategias de la plataforma.

Con esta información, podrás entender mejor cómo ha evolucionado la popularidad de Udemy en términos de suscriptores y tomar decisiones informadas para ajustar el enfoque de negocio en función de los resultados.

```python
# Convertir la columna 'published_timestamp' al formato de fecha
udemy['published_timestamp'] = pd.to_datetime(udemy['published_timestamp'])

# Extraer el año de la columna 'published_timestamp'
udemy['year'] = udemy['published_timestamp'].dt.year

# Obtener el número total de suscriptores por año
subscribers_by_year = udemy.groupby('year')['num_subscribers'].sum()

# Crear el gráfico de líneas para la evolución de suscriptores por año
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot(subscribers_by_year.index, subscribers_by_year.values, marker='o', linestyle='-', color='b')

# Configurar etiquetas y título del gráfico
ax.set_xlabel('Año')
ax.set_ylabel('Número total de suscriptores')
ax.set_title('Evolución de suscriptores en Udemy por año')

# Formatear los valores del eje Y con separadores de miles
ax.yaxis.set_major_formatter(ticker.StrMethodFormatter('{x:,.0f}'))

# Mostrar el gráfico
plt.xticks(subscribers_by_year.index)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd

# Calculando el crecimiento de suscriptores por año
df_udemy['published_timestamp'] = pd.to_datetime(df_udemy['published_timestamp'])
df_udemy['year'] = df_udemy['published_timestamp'].dt.year
df_subscribers_by_year = df_udemy.groupby('year')['num_subscribers'].sum()
growth_rate = (df_subscribers_by_year.max() - df_subscribers_by_year.min()) / df_subscribers_by_year.min() * 100

# Imprimiendo el KPI
print("KPI de crecimiento de suscriptores en Udemy por año:")
print("Crecimiento de suscriptores: {:.2f}%".format(growth_rate))
```
