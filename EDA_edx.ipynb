Cantidad de Inscritos

```python
# Eliminar caracteres no numéricos de la columna 'n_enrolled'
edx['n_enrolled'] = edx['n_enrolled'].str.replace(',', '')
edx['n_enrolled'] = edx['n_enrolled'].str.replace('nan', '')

# Reemplazar valores faltantes con cero
edx['n_enrolled'] = edx['n_enrolled'].fillna(0)

# Convertir la columna 'n_enrolled' en enteros
edx['n_enrolled'] = edx['n_enrolled'].astype(int)

# Ordenar el DataFrame por el número de inscritos en forma descendente
edx_sorted = edx.sort_values('n_enrolled', ascending=False)
```

### KPI edx

KPI de las temáticas más demandadas en edX basado en la cantidad de inscritos en cada temática:

Objetivo: Identificar las temáticas más demandadas en edX con base en la cantidad de inscritos.

Indicador numérico: Número de inscritos por temática.

Plazo: Medir la demanda por temática de forma continua.

Meta: Identificar las 5 temáticas con mayor número de inscritos en edX.

Monitoreo y medición: Analizar regularmente los datos de inscritos por temática y establecer un ranking de las temáticas más demandadas.

```python
import re
import matplotlib.pyplot as plt

# Obtener las 10 temáticas con más suscriptores
top_10_subjects = edx.groupby('subject')['n_enrolled'].sum().nlargest(10)

# Limpiar los valores de la columna 'price' y convertirlos a float
price_values = edx['price'].str.extract(r'(\d+)', expand=False)
edx['price'] = price_values.astype(float)

# Calcular el precio promedio del total de cursos por temática
avg_prices = edx.groupby('subject')['price'].mean()

# Filtrar los datos para las temáticas del top 10
top_10_avg_prices = avg_prices[top_10_subjects.index]

# Crear el gráfico de barras
fig, ax = plt.subplots(figsize=(12, 6))
bar_colors = plt.get_cmap('tab10').colors[:len(top_10_subjects)]
ax.bar(top_10_subjects.index, top_10_subjects, color=bar_colors)

# Configurar el eje y con la función ticker
ax.yaxis.set_major_formatter(ticker.FuncFormatter(lambda x, _: f'{x:,.0f}'))

# Configurar la leyenda con los precios promedio
legend_labels = [f'{subject}\nAvg Price: ${price:.2f}' for subject, price in zip(top_10_subjects.index, top_10_avg_prices)]
ax.legend(legend_labels, loc='upper right')

# Rotar las etiquetas del eje x para una mejor legibilidad
plt.xticks(rotation=45, ha='right')

# Etiquetas y título del gráfico
ax.set_xlabel('Subjects')
ax.set_ylabel('Number of Subscribers')
ax.set_title('Top 10 Subjects by Number of Subscribers')

# Mostrar el gráfico
plt.tight_layout()
plt.show()
```
